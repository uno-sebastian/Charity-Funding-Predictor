```python
import numpy as np
print('numpy: %s' % np.__version__)

import pandas as pd
print('pandas: %s' % pd.__version__)

import tensorflow as tf
print('tensorflow: %s' % tf.__version__)

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
print('sklearn: %s' % sklearn.__version__)

import matplotlib
import matplotlib.pyplot as plt
print('matplotlib: %s' % matplotlib.__version__)
```

```python
# Display all the columns for the dataframes (not-truncated)
pd.set_option("display.max_columns", None)
```

## Load the CSV

```python
#  Import and read the charity_data.csv
application_df = pd.read_csv("Resources/clean_charity_data.csv")
application_df
```

* **APPLICATION_TYPE**â€”Alphabet Soup application type
* **AFFILIATION**â€”Affiliated sector of industry
* **CLASSIFICATION**â€”Government organization classification
* **USE_CASE**â€”Use case for funding
* **ORGANIZATION**â€”Organization type
* **STATUS**â€”Active status
* **INCOME_AMT**â€”Income classification
* **SPECIAL_CONSIDERATIONS**â€”Special consideration for application
* **ASK_AMT**â€”Funding amount requested
* **IS_SUCCESSFUL**â€”Was the money used effectively

```python
# Split the columns into quantitative and categorical
all_quantitative_columns = set(application_df.describe().columns)
all_categorical_columns = set(application_df.columns) - all_quantitative_columns
print("quantitative columns\n\t" + "\n\t".join(all_quantitative_columns))
print()
print("categorical columns\n\t" + "\n\t".join(all_categorical_columns))
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, columns=all_categorical_columns)
application_df
```

```python
# shuffle our data to remove any potential bias
application_df = sklearn.utils.shuffle(application_df)
```

```python
# Split our preprocessed data into our features and target arrays
target_feature = "IS_SUCCESSFUL"

X = application_df.drop(columns=[target_feature])
y = application_df[target_feature].values.ravel()
print(X.shape, y.shape)
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y)
print("train\t", X_train.shape, y_train.shape)
print("test\t", X_test.shape, y_test.shape)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_shape=(len(X_train_scaled[0]),)
hidden_layers = [
    80, 
    30, 
    1,
]

nn = tf.keras.models.Sequential()
nn.add(tf.keras.Input(shape=input_shape))
for layer in hidden_layers:
    nn.add(tf.keras.layers.Dense(layer))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(
    optimizer = "Adam", 
    loss = "mse", 
    metrics = ["mae"]
)
```

```python
# Train the model
history = nn.fit(
    X_train_scaled, 
    y_train, 
    epochs = 100
)
history.params
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```

## Load the saved model

```python
loaded_model = tf.keras.models.load_model("AlphabetSoupCharity_Optimization.h5")
# Check its architecture
loaded_model.summary()
```
